思路：

先把数组整体反转。

再把前 k 个数字反转。

再把后 n-k 个数字反转。



```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        def reverse(i:int , j:int) -> None:
            while i < j:
                nums[j], nums[i] = nums[i] , nums[j]
                i += 1
                j -= 1

        n = len(nums)
        k %= n
        reverse(0 , n-1)
        reverse(0 , k-1)
        reverse(k , n-1)
```
